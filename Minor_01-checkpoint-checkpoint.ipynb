```python
items= pd.DataFrame(data, columns = ['Item_Type', 'Item_Name', 'Item_Visibility', 'Item_MRP', 'Discount', 'Final_MRP'])
items.sort_values('Item_Name')
```

```python
a = ["Brownies","Chocolate cake", "Coconut", "Old Monk", "Red Bliss", "Ketel one", "Fruit cake", "Palm Dryfruit Cake"]
#a = ["HD05", "HH02", "HH05", "HE05", "HE15", "HE09"]'
```

```python
buy=data.query('Item_Name==@a')
buy.sort_values('Item_Name')
```

```python
buy.info()
```

```python
total_buy_store=buy.groupby('Store_Id')['Final_MRP'].sum().sort_values()
total_buy_store
```

```python
#plt.figure(figsize=(8,4))
#tbuy.plot(kind='line', marker='o', markersize=7)
#plt.plot(total_buy_store)
```

```python
#change background color of graph to yellow
sns.set_style('darkgrid', {'axes.facecolor': 'white', 'figure.facecolor': 'white', 'figure.edgecolor': 'black', 'axes.edgecolor': 'black',})

plt.figure(figsize=(20,10))
plt.plot(total_buy_store, marker='o', linewidth=3, markersize=8, color='brown', label='Total Sales')
plt.legend( loc='upper left', shadow=True)
plt.title('Total Sales per Store', fontsize=20)
plt.xlabel('Store Id', fontsize=20)
plt.ylabel('Total Sales', fontsize=20)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5, which = 'both', zorder = 0, clip_on = False,)
# show y-values on the line of plot
for x,y in zip(total_buy_store.index, total_buy_store.values):
    label = "{:.2f}".format(y)
    plt.annotate(label, (x,y), textcoords="offset points", xytext=(0,10), ha='center', color='black', fontsize=14)
plt.show()
```

```python
minimum_price=total_buy_store.idxmin()
minimum_price
```

```python
selected_store=data[data['Store_Id']==minimum_price]
selected_store.sort_values('Off_Price', ascending=False)
```

```python
total_buy_rating=buy.groupby('Store_Id')['Rating'].sum().sort_values(ascending=False)
total_buy_rating
```

```python
#plt.figure(figsize=(8,4))
#tbuy.plot(kind='line', marker='o', markersize=7)
#plt.plot(total_buy_rating)
```

```python
#change background color of graph to yellow
sns.set_style('darkgrid', {'axes.facecolor': 'white', 'figure.facecolor': 'white', 'figure.edgecolor': 'black', 'axes.edgecolor': 'black',})

plt.figure(figsize=(20,10))
plt.plot(total_buy_rating, marker='o', linewidth=3, markersize=8, color='brown', label='Total Sales')
plt.legend( loc='upper left', shadow=True)
plt.title('Total Sales per Store', fontsize=20)
plt.xlabel('Store Id', fontsize=20)
plt.ylabel('Total Sales', fontsize=20)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5, which = 'both', zorder = 0, clip_on = False,)
# show y-values on the line of plot
for x,y in zip(total_buy_rating.index, total_buy_rating.values):
    label = "{:.2f}".format(y)
    plt.annotate(label, (x,y), textcoords="offset points", xytext=(0,10), ha='center', color='black', fontsize=14)
plt.show()
```
